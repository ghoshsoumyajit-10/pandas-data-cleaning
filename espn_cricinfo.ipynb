```python
import pandas as pd
```

```python
df=pd.read_csv("espn.csv")
print(df)
print(f"{df.shape[0]} rows and {df.shape[1]} columns")
               
```

```python
df.columns=df.columns.str.strip()
print(df.columns.to_list())
```

```python
df=df.rename(columns={"NO":"Not_Out","HS":"Highest_Score","BF":"Balls_Faced","SR":"Strike_Rate"})
print(df.head())
```

```python
#check null values
df.isnull().any()
```

```python
df[df["Balls_Faced"].isna()==True]
```

```python
df[df["Strike_Rate"].isna()==True]
```

```python
df[df["0"].isna()==1]
```

```python
df["Balls_Faced"]=df["Balls_Faced"].fillna(0)
df["Strike_Rate"]=df["Strike_Rate"].fillna(0)
df["0"]=df["0"].fillna(0)
df.isnull().any()
```

```python
df[df["Player"]=="ED Weekes (WI)"]
df[df["Player"].isin(["Hashmatullah Shahidi (AFG)","GA Headley (WI)"])]
```

```python
#can be used to drop duplicates
df[df["Player"].duplicated()==1]
df=df.drop_duplicates()
```

```python
#Split span into start and end date
df["Debut_Year"]=df["Span"].str.split('-').str[0]
df["Retirement_Year"]=df["Span"].str.split('-').str[1]
df.drop("Span",axis=1,inplace=True)
print(df)
```

```python
#Split into player and country name
df["Country"]=df["Player"].astype(str).str.split(r"\(").str[1]
df["Country"]=df["Country"].astype(str).str.split(r"\)").str[0]
df.head()

```

```python
df["Player"] = df["Player"].astype(str).str.split(r"\(").str[0]
print(df)
```

```python
#change data types
df.dtypes
```

```python
df["Highest_Score"]=df["Highest_Score"].str.split("*").str[0]
df.head()
df["Highest_Score"]=df["Highest_Score"].fillna(0).astype(int)
df.head()

```

```python
df.dtypes
```

```python
df=df.astype({"Debut_Year":int,"Retirement_Year":int})

```

```python
df["Mat"]=df["Mat"].str.split("*").str[0]
df["Mat"]=df["Mat"].fillna(0).astype(int)
```

```python
df["Balls_Faced"] = (
    df["Balls_Faced"]
    .astype(str)
    .str.split("+")
    .str[0]
    .replace("nan", pd.NA)
)
```

```python
df["Balls_Faced"]=df["Balls_Faced"].fillna(0).astype(int)

```

```python
df["Career_Length"]=df["Retirement_Year"]-df["Debut_Year"]
print(df.head())
df.index=df.index+1
```

```python
#Average career length of Cricketers
df["Career_Length"].mean()
```

```python
#Avg Batting Strike Rate for cricketers who played for more than 10 years
df[df["Career_Length"]>10]["Strike_Rate"].mean()
```

```python
#Count of players who played before 1960
df[df["Debut_Year"]<1960]["Player"].count()
```

```python
#Max highest Innings score by Country
df.groupby("Country")["Highest_Score"].max().sort_values(ascending=False)
```

```python
#Hundreds,fifties and ducks(0) Avg by country
df.groupby("Country")[["100","50","0"]].mean()
```
